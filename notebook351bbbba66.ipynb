```python
df = pd.read_csv(r"/kaggle/input/online-retail-dataset/online_retail.csv")
df
```

```python
df.info()
```

```python
df['CustomerID'].isnull().value_counts()
# Means there are 135080 values missing in CustomerID
```

```python
# Fill the missing values of CustomerID with Unknown
df['CustomerID'] = df['CustomerID'].fillna('Unknown')
df['CustomerID'].info() # Now missing values is filled by Unknown customer 
```

```python
df['Total Revenue'] = df['Quantity']*df['UnitPrice']
df.head()
```

```python
df['InvoiceDate'].info()
```

```python
# Changing dtype of InvoiceDate from object to datatime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceDate'].info()
```

```python
# Total units purchased for each Products and their total revenue
total_units = df.groupby('Description')[['Quantity','Total Revenue']].sum().sort_values(by = 'Total Revenue', ascending = False)
total_units
```

```python
# Removing time from Invoice Date
df['Date'] = df['InvoiceDate'].dt.date
df.head()
```

```python
# Changing dtype of date from object to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
# Total Revenue on each day
day_wise = df.groupby('Date')['Total Revenue'].sum()
day_wise
```

```python
# Total Revenue by Countrywise
country_wise = df.groupby('Country')['Total Revenue'].sum()
country_wise
```
